```python
import tkinter as tk
from googletrans import Translator 


class TranslationApp:
    def __init__(self, master):
        self.master = master #mainwindow
        self.master.title("LanguageTranlationApp")
        self.input_label = tk.Label(master, text = "Enter text to translate")
        self.input_label.pack(pady = 5)
        self.input_text = tk.Text(master, height = 5, width = 50)
        self.input_text.pack(padx=10, pady = 5)
        self.translate_button = tk.Button(master, text = "Transalte", command = self.translate_text)
        self.translate_button.pack(pady=5)
        
        self.output_label = tk.Label(master, text="Translation:")
        self.output_label.pack(padx = 10, pady = 5)
        self.output_text = tk.Text(master, height = 5, width = 50, state = tk.DISABLED)
        self.output_text.pack(padx=10, pady=5)
        
        
    def translate_text(self):
        input_text = self.input_text.get("1.0", tk.END).strip()

        if not input_text: 
            tk.messagebox.showwarning("Warning", "Please enter something")
            return

        translater = Translator()
        translation = translater.translate(input_text, dest ='fr').text

        self.output_text.config(state = tk.NORMAL)
        self.output_text.delete("1.0", tk.END)
        self.output_text.insert(tk.END, translation)
        self.output_text.config(state = tk.DISABLED)
    
def main():
    root = tk.Tk()
    app = TranslationApp(root)
    root.mainloop()

        
        
```

```python
if __name__ == "__main__":
    main()
```
