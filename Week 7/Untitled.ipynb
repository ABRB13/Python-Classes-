```python
import turtle 
import random 

screen = turtle.Screen()
screen.title("Racing Game")
screen.bgcolor('green')
screen.setup(width = 600, height = 400)

f_line = turtle.Turtle()
f_line.penup()
f_line.goto(250,180)
f_line.pendown()
f_line.goto(250,-180)
f_line.hideturtle()

colors = ['red', 'orange', 'blue', 'pink', 'yellow']

players = len(colors)
turtles = []


for i in range(players)
    player = turtle.Turtle()
    player.shape('turtle')
    player.color(colors[i])
    player.penup()
    player.goto(-250, -150 + i * 60)
    turtles.append(player)
    
def race(t):
    distance = random.randint(1,10)
    t.forward(distance)
    
# for t in turtles:
#     screen.listen()

winner = None 

while True:
    for t in turtles:
        race(t)
        if t.xcor() >= 230:
            winner = t.color()[0]
            break 
    if winner:
        break 

text = turtle.Turtle()

text.penup()
text.color('black')
text.goto(250,150)
text.write(f"The {winner} turtle wins", align = 'center', font=("Arial", 20, "bold"))
text.hideturtle()

turtle.done()
    
    
```

```python
import turtle 


screen = turtle.Screen()
screen.title("Patterns")
screen.bgcolor('white')
screen.setup(width = 600, height = 400)

pen = turtle.Turtle()

pen.speed(2)
pen.width(2)
pen.color('blue')

def square():
    for i in range(4):
        pen.forward(100)
        pen.right(90)
        
def tri():
    for i in range(3):
        pen.forward(100)
        pen.left(130)

pen.penup()
pen.goto(-150,100)
pen.pendown()
square()
tri()

turtle.done()
        
```

```python
#custome turtle

import math 

class Turtle:
    def __init__(self,name,color):
        self.name = name
        self.color = color
        self.position = (0,0)
        self.angle = 0
        print(name, color)
        
    def move_forward(self, distance):
        new_x = self.position[0] + distance * math.cos(math.radians(self.angle))
        new_y = self.position[1] + distance * math.sin(math.radians(self.angle))
        self.position = (new_x, new_y)
        print(f"{self.name} moved forward to {self.position}")
    
    def turn_right(self,angle):
        self.angle -= angle
        
    
        
        
        
        
```
