```python
class BankAccount:
    def __init__(self, number, balance):
        self.number = number 
        self.balance = balance 
    
        
    def get_balance(self):
        return self.balance
        
    def deposit(self, amount):
        self.balance += amount 
        
    
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount 
        else:
            print("insufficient")
        
```

```python
account1 = BankAccount("123", 500)
```

```python
print(account1.get_balance())
```

```python
print(account2.get_balance())
```

```python
account2 = BankAccount("234", 1000)
```

```python
account1.withdraw(200)
```

```python
class shape:
    def area(self):
        return "Area Calculation"    
class Circle(shape):
    def __init__(self, radius):
        self.radius = radius
        
    def area(self):
        return 3.14 * self.radius**2
    
class rect(shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width 
    def area(self):
        return(self.length * self.width)
```

```python
s = shape()
```

```python
c = Circle(5)
```

```python
r = rect(4,6)
```

```python
s.area()
```

```python
c.area()
```

```python
r.area()
```

```python
#widgets 

import tkinter as tk
from tkinter import ttk 

root = tk.Tk()
root.title("window")
root.geometry("600x400")

email = tk.StringVar()
password = tk.StringVar()


def login_clicked():
    #callback function 
    
    msg = f"you entered email: {email.get()} and password: {password.get()}"
    
    info(title = "information",
            message = msg)
    
signin = ttk.Frame(root)
signin.pack(padx=10, pady=10, fill = 'x', expand = True)

#email

email_label = ttk.Label(signin, text = "Address: ")
email_label.pack(fill = 'x', expand = True)

email_entry = ttk.Entry(signin, textvariable = email)

email_entry.pack(fill = 'x', expand = True)

password_label = ttk.Label(signin, text = 'Password: ')
password_entry = ttk.Entry(signin, textvariable = password)

password_label.pack(fill = 'x', expand = True)

password_entry.pack(fill = 'x', expand = True)

login_button = ttk.Button(signin, text = "Login", command = login_clicked)
login_button.pack(fill = 'x', expand = True)

root.mainloop()


#pack, grid, place






```

```python
import tkinter as tk

from tkinter import Menu 

```

```python
root = tk.Tk()
root.title()

menubar = Menu(root)
root.config(menu=menubar)

file_menu = Menu(menubar)

file_menu.add_command(label = 'New')
file_menu.add_command(label = 'Open')
file_menu.add_command(label = 'Save')

file_menu.add_command(label = 'Exit', 
                     command = root.destroy)

sub_menu = Menu(file_menu)
sub_menu.add_command(label = "Shortcut")

file_menu.add_cascade(label = 'Pref', menu = sub_menu)


menubar.add_cascade(label = 'File', menu = file_menu)

root.mainloop()







```
